```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv", index_col=0)
df
```

#columns Remove


```python
df.drop(['School Level','Year','Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 80 or Above','Percent Scoring 80 or Above','Number Scoring CR','Percent Scoring CR'], axis=1, inplace=True)

```

```python
df
```

#cleaning by removing letters where the numbers should be


```python
df[df['Number Scoring 65 or Above'].str.contains('0')]
```

#remove duplicates

```python
duplicates=df.duplicated(subset=None, keep='first')
print(duplicates)
```

```python
df.drop_duplicates(subset=None, keep='first', inplace=True)

```

#visualization by regents exam

```python
df.groupby(['Regents Exam'])['Number Scoring 65 or Above'].count().sort_values(ascending = False).plot(kind='bar')
```

#i wanted a total of all the regents exams in new york

```python
df_count= df.groupby(['Regents Exam'])['Number Scoring 65 or Above'].count()
df_count
```
